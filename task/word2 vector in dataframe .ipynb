```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
```

```python
import pandas as pd
dataset=pd.read_csv(r"D:\housing_dataset.csv")

dataset.head()
```

```python
dataset=dataset.iloc[:500,4:6]
```

```python
import re
```

```python
dataset = dataset.replace(r"<{\s*(.*?)\s*}>", r"\1", regex=True)
```

```python
dataset.head(2)
```

```python
data=dataset.apply(lambda x:x.str.split())
```

```python
data.head()
```

```python
flatten_data=data.values.flatten()
flatten_data
```

```python
model = Word2Vec(flatten_data, vector_size=10, window=5, min_count=1, workers=4)
model.wv
```

```python
import numpy as np
```

```python


# Ensure the Word2Vec model is loaded before calling this function
def return_vector(sentence, model):
    vector = []
    for word in sentence.split():  # Fixed variable name
        if word in model.wv:
            vector.append(model.wv[word])
    if vector:
        return np.mean(vector, axis=0)  
    else:
        return np.zeros(model.vector_size)  # Fixed issue

    
```

```python
vectorized_data={}
```

```python
for col in dataset.columns:
    vectorized_data[col] = dataset[col].apply(lambda x: return_vector(x, model)).tolist()
```

```python
df=pd.DataFrame(vectorized_data)
df.head(2)
```

```python
model.build_vocab(flatten_data,progress_per=1000)
```

```python
model.train(flatten_data,total_examples=model.corpus_count,epochs=model.epochs)
```
